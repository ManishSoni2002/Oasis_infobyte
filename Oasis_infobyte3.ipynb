Firstly create model for data with outliers

### Identifying independent variable and dependent variable

```python
y=num_cols[['price']]                              #Dependent variable
x=num_cols.drop(columns=['car_ID','price'])
```

### Train and test split 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=333)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

we are using standardization because data have outliers.                                   

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler() # creating an instance of the class object
scaled_x_train=sc.fit_transform(x_train)
scaled_x_test=sc.fit_transform(x_test)
```

```python
df_scaled_x_train=pd.DataFrame(scaled_x_train,columns=x_train.columns)
df_scaled_x_train.head()
```

```python
df_scaled_x_test=pd.DataFrame(scaled_x_test,columns=x_test.columns)
df_scaled_x_test.head()
```

### Creating a model

```python
from sklearn.linear_model import LinearRegression
MLR=LinearRegression()
mlr_model=MLR.fit(scaled_x_train,y_train)
```

### Find slope ,intercept

```python
print("Slope:",mlr_model.coef_)
print("Intercept:",mlr_model.intercept_)
```

### R square value

```python
print("R square value:",mlr_model.score(df_scaled_x_train,y_train))
```

### Predict the values on test data using mlr model

```python
y_test['mlr_pred']=mlr_model.predict(df_scaled_x_test)
y_test['mlr_pred']=round(y_test['mlr_pred'],3)
y_test.head()
```

```python
y_test[['price','mlr_pred']].plot()
```

### Calculate RMSE value

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

error=y_test['price']-y_test['mlr_pred']
sqrt(mean_squared_error(y_test['price'],y_test['mlr_pred']))
```

### Model evalution

```python
sns.distplot(error)
```

# Ridge regression

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=50,max_iter=100,tol=0.1)
ridge_model=ridge.fit(df_scaled_x_train,y_train)
ridge_model
```

### R square value for ridge value

```python
print("R square value for train data:",ridge_model.score(df_scaled_x_train,y_train))
print("R square value for test data:",ridge_model.score(df_scaled_x_test,y_test['price']))
```

### Predict price using ridge model

```python
y_test['ridge_pred']=ridge_model.predict(df_scaled_x_test)
y_test['ridge_pred']=round(y_test['ridge_pred'],3)
y_test.head()
```

### RMSE value for Ridge model

```python
error=y_test['price']-y_test['ridge_pred']
sqrt(mean_squared_error(y_test['price'],y_test['ridge_pred']))
```

# Lasso Regression model

```python
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=50,max_iter=100,tol=0.1)
lasso_model=lasso.fit(df_scaled_x_train,y_train)
lasso_model
```

### R square value for lasso model

```python
print("R square value for train data:",lasso_model.score(df_scaled_x_train,y_train))
print("R square value for test data:",lasso_model.score(df_scaled_x_test,y_test['price']))
```

### Predict price using Lasso model

```python
y_test['lasso_pred']=lasso_model.predict(df_scaled_x_test)
y_test['lasso_pred']=round(y_test['lasso_pred'],3)
y_test.head()
```

### RMSE value for Lasso model

```python
error=y_test['price']-y_test['lasso_pred']
sqrt(mean_squared_error(y_test['price'],y_test['lasso_pred']))
```
